```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 
```

```python
df = pd.read_excel('업종별경기동향.xlsx')
df
```

```python
sorted = df.sort_values(by='2021.02')
sorted
```

```python
sorted = df.sort_values(by='2022.02')
sorted
```

```python
sum = sorted.groupby('업종코드별').sum()
sum
```

```python
sum.plot()
plt.title("Overall performance trend of sum")
plt.xlabel("Graph")
plt.ylabel("Data")
plt.show()
```

```python
df = df[['업종코드별','BSI코드별', '2018.02', '2019.02', '2020.02', '2021.02', '2022.02']]
df
```

```python
mean = sorted.groupby('업종코드별').mean()
mean
```

```python

plt.figure(figsize = (10, 8))


for name, group in df.groupby('BSI코드별'):
       
       plt.scatter(group['2018.02'], group['2019.02'], label = name)


plt.xlabel('2018.02')
plt.ylabel('2019.02')


plt.legend()

plt.show()
```
