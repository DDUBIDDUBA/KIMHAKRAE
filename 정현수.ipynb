# 업종별 창업 기업수 규모 변화

# 업종별 창업 기업수 변화

2018년부터 2022년까지의 업종별 창업 기업수의 변화를 시각화하기 위한 단계이다.

```python
plt.plot(df1t.iloc[:, [0]], color = 'orangered')
plt.title("년도별 '농업, 임업 및 어업 및 광업'의 창업 기업수 변화", fontsize = 20)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.show()
```

```python
plt.plot(df1t.iloc[:, [1]], color = 'orange')
plt.title("년도별 '제조업'의 창업 기업수 변화", fontsize = 20)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.show()
```

```python
plt.plot(df1t.iloc[:, [2]], color = 'limegreen')
plt.title("년도별 '전기, 가스, 증기 및 공기 조절 공급업'의 창업 기업수 변화", fontsize = 20)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.show()
```

```python
plt.plot(df1t.iloc[:, [3]], color = 'dodgerblue')
plt.title("년도별 '건설업'의 창업 기업수 변화", fontsize = 20)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.show()
```

```python
plt.plot(df1t.iloc[:, [4]], color = 'darkslateblue')
plt.title("년도별 '서비스업'의 창업 기업수 변화", fontsize = 20)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.show()
```

2018년부터 2022년까지의 각 업종별 창업 기업수의 변화를 나타낸 그래프들을 하나의 그래프로 묶어보았다.

```python
plt.plot(df1t.iloc[:, [0]], color = 'orangered')
plt.plot(df1t.iloc[:, [1]], color = 'orange')
plt.plot(df1t.iloc[:, [2]], color = 'limegreen')
plt.plot(df1t.iloc[:, [3]], color = 'dodgerblue')
plt.plot(df1t.iloc[:, [4]], color = 'darkslateblue')

plt.title('년도별 업종별 창업 기업수 변화', fontsize = 20)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.legend(df1t.columns)

plt.show()
```

서비스업과 다른 업종의 규모 차이가 커서
서비스업을 제외한 나머지 업종의 그래프들을 함께 묶어보았다.

```python
plt.plot(df1t.iloc[:, [0]], color = 'orangered')
plt.plot(df1t.iloc[:, [1]], color = 'orange')
plt.plot(df1t.iloc[:, [2]], color = 'limegreen')
plt.plot(df1t.iloc[:, [3]], color = 'dodgerblue')

plt.title('년도별 업종별 창업 기업수 변화', fontsize = 20)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.legend(df1t.columns)

plt.show()
```

년도별 창업 기업수는 건설업과 농업, 임업 및 어업 및 광업은 꾸준히 증가하는 추세를, 제조업과 전기, 가스, 증기 및 공기 조절 공급업은 꾸준히 증가하는 추세를 보인다는 것을 확인할 수 있다. 하지만 업종별로 산업에서 담당하는 규모의 변동은 크지 않다는 것을 확인할 수 있다.

# 제조업의 세부 업종별 규모 변화

2018년부터 2022년까지의 제조업의 세부 업종별 규모 변화를 시각화하기 위한 단계이다.

```python
make = df.loc[[5, 6, 7, 8, 9, 10, 11, 12, 13]]
make
```

```python
make.drop('업종별(1)', axis = 1, inplace = True)
make.sort_values(by = '2018.02', ascending = False, inplace = True)
make
```

```python
mt = make.transpose()
mt
```

```python
mt.columns = mt.loc['업종별(2)']
mt.drop('업종별(2)', inplace = True)
mt.columns = list(mt.columns)
mt
```

2018년부터 2022년까지의 제조업의 세부 업종별 규모 변화를 나타낸 그래프이다.

```python
mt.plot(kind = 'bar', color = ['red', 'darkorange', 'gold', 'yellowgreen', 'forestgreen', 'deepskyblue', 'dodgerblue', 'steelblue', 'slateblue'], stacked = True)
```

제조업의 세부 업종별 규모 중 가장 큰 규모를 가지는 업종은 기계, 금속이다.
제조업의 세부 업종별 규모 중 유의미한 변화를 보이는 업종은 기타제조업이다. 기타제조업은 증가하는 추세를 보이다가 2020년 급격히 감소하면서, 그 규모가 꾸준히 감소하고 있다.

# 서비스업의 세부 업종별 규모 변화
